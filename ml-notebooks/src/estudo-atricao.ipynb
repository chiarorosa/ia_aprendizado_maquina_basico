```python
df_sem_nulos = df.dropna()
print(df_sem_nulos.isna().sum(), df_sem_nulos['Attrition'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df_sem_nulos.copy() # Crie uma cÃ³pia do DataFrame

colunasCategoricas = ['Attrition', 'BusinessTravel', 'Department', 
                      'EducationField', 'Gender', 'JobRole', 'MaritalStatus']  # Substitua pelos nomes das colunas categÃ³ricas em seus dados
label_encoder = LabelEncoder() # Inicialize o LabelEncoder

# Aplique o LabelEncoder Ã s colunas categÃ³ricas nos dados de treinamento e teste
for col in colunasCategoricas:
    df_encoded[col] = label_encoder.fit_transform(df_encoded[col])
```

```python
X = df_encoded.drop("Attrition", axis=1) # Todas as colunas, exceto Classe
y = df_encoded['Attrition'] # Apenas a coluna Classe
```

```python
X.iloc[0:10]
```

```python
from imblearn.over_sampling import RandomOverSampler
ros =  RandomOverSampler(sampling_strategy=1)

X_resample, y_resample = ros.fit_resample(X, y)

contagemPositivo = len(y_resample[y_resample==1])
contagemNegativo = len(y_resample[y_resample==0])

print("AtriÃ§Ã£o negativo: {:.2f}%".format(
    (contagemNegativo / (len(y_resample))*100)
    ))
print("AtriÃ§Ã£o positivo: {:.2f}%".format(
    (contagemPositivo / (len(y_resample))*100)
    ))
```
